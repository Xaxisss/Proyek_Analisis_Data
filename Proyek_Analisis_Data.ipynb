### Cleaning Data

```python
geolocation_df.drop_duplicates(inplace=True)
```

```python
geolocation_df.duplicated().sum()
```

```python
order_items_df['shipping_limit_date'] = pd.to_datetime(order_items_df['shipping_limit_date'])
```

```python
order_items_df.info()
```

```python
columns = ['review_creation_date', 'review_answer_timestamp']

for column in columns:
    order_reviews_df[column] = pd.to_datetime(order_reviews_df[column])
```

```python
order_reviews_df.info()
```

Dalam mengatasi Missing value, kita harus berhati hati, pada kasus ini terjadi dalam dataset order_reviews, yaitu dataset yang menunjukkan feedback dari customer, tidak heran jika bagian review_comment_title, dan review_comment_message terdapat missing value, karena mungkin saja banyak pelanggan yang hanya me review score dan tidak mengisi komen reviewnya. Maka dari itu pada kasus ini, kita akan mengisi missing value menggunakan metode imputasi dengan ketentuan sebagai berikut:
- review_comment_title diisi dengan string kosong("")
- review_comment_message diisi dengan pesan "No Comment"


```python
order_reviews_df.loc[:, "review_comment_title"] = order_reviews_df["review_comment_title"].fillna("")
```

```python
order_reviews_df.loc[:, "review_comment_message"] = order_reviews_df["review_comment_message"].fillna("No comment")
```

```python
order_reviews_df.isna().sum()
```

dalam menangani orders_df yang mempunyai missing value juga harus berhati hati, apalagi data kali ini berkaitan dengan format datetime sehingga tidak bisa diisi dengan string atau yang lain, bila kita lihat terdapat 3 kolom dengan missing valud dengan ketentuan sebagai berikut:

order_approved_at                 160
order_delivered_carrier_date     1783
order_delivered_customer_date    2965

karena data yang kita punya banyak, yaitu 99.441 data, maka kolom order_approved_at, dan order_delivery_carrier_date yang mempunyai NaN akan kita hapus, karena cenderung sedikit dibanding data, akan tetapi kolom order_delivery_customer_date, kita akan gunakan metode imputation dengan kolom order_delivered_customer_date, karena kita berasumsi bahwa customer sudah menerima barangnya, sesuai dengan tanggal estimasi, hanya saja belum menekan tombol konfirmasi.

```python
orders_df.dropna(subset=["order_approved_at", "order_delivered_carrier_date"], inplace=True)
```

```python
orders_df.loc[:, "order_delivered_customer_date"] = orders_df["order_delivered_customer_date"].fillna(
    orders_df["order_estimated_delivery_date"]
)

```

```python
orders_df.isna().sum()
```

```python
orders_df.info()
```

```python
date_column = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']
for column in date_column:
    orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
orders_df.info()
```

Sekali lagi, untuk menangani missing value dari products_df tidak boleh asal, pada product_category_name nilai yang hilang sekitar 1.85%, dan untuk menangani product_category_name lebih baik kita isi dengan "Unknown", lalu untuk product_name_lenght, product_description_lenght, product_photos_qty akan kita isi dengan nilai median, karena nilai median cenderung stabil terhadap outlier.sedangkan sisanya akan kita hapus, karena tidak signifikan terhadap jumlah data, hanya sekitar 0.006%

```python
products_df.loc[:,"product_category_name"] = products_df["product_category_name"].fillna("Unknown")
```

```python
misscol = ['product_name_lenght', 'product_description_lenght', 'product_photos_qty']

for column in misscol:
    products_df[column] = products_df[column].fillna(products_df[column].median())
```

```python
products_df.dropna(subset=["product_weight_g", "product_length_cm", "product_height_cm", "product_width_cm"], inplace=True)
```

```python
products_df.isna().sum()
```

```python
wrongcolumn = ['product_name_lenght', 'product_description_lenght', 'product_photos_qty']

for column in wrongcolumn:
    products_df[column] = products_df[column].astype(int)
```

```python
products_df.info()
```

```python
products_df[products_df["product_weight_g"] == 0].shape[0]
```

```python
#Karena jumlahnya hanya 4, akan kita hapus
products_df = products_df[products_df["product_weight_g"] > 0]
```

## Analisis Lanjutan (Opsional) : Analisis RFM dan pengaruhnya terhadap Churn Rate

```python
# Pastikan 'order_purchase_timestamp' dalam format datetime
main_customers_df["order_purchase_timestamp"] = pd.to_datetime(main_customers_df["order_purchase_timestamp"])

# Pilih kolom yang diperlukan
rfm_df = main_customers_df[["customer_unique_id", "order_id", "order_purchase_timestamp", "price"]]
```

```python
# Tentukan tanggal analisis (misalnya, hari terakhir transaksi)
reference_date = main_customers_df["order_purchase_timestamp"].max()

# Hitung RFM Metrics
rfm = rfm_df.groupby("customer_unique_id").agg({
    "order_purchase_timestamp": lambda x: (reference_date - x.max()).days,  # Recency
    "order_id": "count",  # Frequency
    "price": "sum"  # Monetary
}).reset_index()

# Ubah nama kolom
rfm.columns = ["customer_unique_id", "Recency", "Frequency", "Monetary"]
```

```python
print(rfm.dtypes)
```

```python
rfm["Recency"] = pd.to_numeric(rfm["Recency"], errors="coerce")
rfm["Frequency"] = pd.to_numeric(rfm["Frequency"], errors="coerce")
rfm["Monetary"] = pd.to_numeric(rfm["Monetary"], errors="coerce")
```

```python
print(rfm.isnull().sum())
```

```python
rfm = rfm.drop(columns=["customer_unique_id"])
```

```python
start_date = main_customers_df["order_purchase_timestamp"].min()
end_date = main_customers_df["order_purchase_timestamp"].max()

print(f"Periode transaksi: {start_date} - {end_date}")
```

```python
# Fungsi untuk memberi skor
# Fungsi untuk memberi skor Recency (semakin kecil lebih baik)
def r_score(x, quantiles):
    if x <= quantiles[0.2]: return 5
    elif x <= quantiles[0.4]: return 4
    elif x <= quantiles[0.6]: return 3
    elif x <= quantiles[0.8]: return 2
    else: return 1

# Fungsi untuk memberi skor Frequency dan Monetary (semakin besar lebih baik)
def fm_score(x, quantiles):
    if x <= quantiles[0.2]: return 1
    elif x <= quantiles[0.4]: return 2
    elif x <= quantiles[0.6]: return 3
    elif x <= quantiles[0.8]: return 4
    else: return 5

# Hitung kuantil untuk setiap metrik
quantiles = rfm.quantile(q=[0.2, 0.4, 0.6, 0.8]).to_dict()

# Terapkan skor dengan aturan yang benar
rfm["R_Score"] = rfm["Recency"].apply(lambda x: r_score(x, quantiles["Recency"]))
rfm["F_Score"] = rfm["Frequency"].apply(lambda x: fm_score(x, quantiles["Frequency"]))
rfm["M_Score"] = rfm["Monetary"].apply(lambda x: fm_score(x, quantiles["Monetary"]))

# Gabungkan skor
rfm["RFM_Score"] = (
    rfm["R_Score"].astype(str) + 
    rfm["F_Score"].astype(str) + 
    rfm["M_Score"].astype(str)
)

```

```python
# Batas churn (misalnya 180 hari sejak transaksi terakhir)
churn_threshold = 180

# Menambahkan kolom Churn (1 jika churn, 0 jika tidak)
rfm["Churn"] = np.where(rfm["Recency"] > churn_threshold, 1, 0)
```

```python
# Rata-rata Recency antara pelanggan churn vs non-churn
rfm.groupby("Churn")["Recency"].mean()

# Rata-rata Frequency antara pelanggan churn vs non-churn
rfm.groupby("Churn")["Frequency"].mean()

# Rata-rata Monetary antara pelanggan churn vs non-churn
rfm.groupby("Churn")["Monetary"].mean()
```

```python
plt.figure(figsize=(12, 5))
sns.histplot(rfm["RFM_Score"], bins=10, kde=True)
plt.title("Distribusi RFM Score", fontsize=15)
plt.xlabel("RFM Score")
plt.ylabel("Jumlah Pelanggan")
labels = plt.xticks()[0]  
plt.xticks(labels[::2], rotation=45)  

plt.show()

```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x="Churn", y="Recency", data=rfm)
plt.title("Recency antara Pelanggan Churn dan Non-Churn")
plt.xlabel("Churn (1 = Churn, 0 = Tidak Churn)")
plt.ylabel("Recency (Hari sejak Pembelian Terakhir)")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
rfm_churn_rate = rfm.groupby("RFM_Score")["Churn"].mean()
rfm_churn_rate = rfm_churn_rate.sort_values(ascending=False)
sns.barplot(y=rfm_churn_rate.values, x=rfm_churn_rate.index, palette="coolwarm")
plt.xticks(rotation=90)
plt.title("Churn Rate berdasarkan RFM Score")
plt.xlabel("RFM Score")
plt.ylabel("Churn Rate")
plt.show()
```

```python
# Hitung rata-rata dari setiap metrik RFM
mean_recency = rfm["Recency"].mean()
mean_frequency = rfm["Frequency"].mean()
mean_monetary = rfm["Monetary"].mean()

print(f"Mean Recency: {mean_recency}, Mean Frequency: {mean_frequency}, Mean Monetary: {mean_monetary}")
```
